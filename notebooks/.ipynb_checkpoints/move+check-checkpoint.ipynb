```python
new = {'num_awds': []}
```

```python
new['num_awds']=[]
rawdir= '/data/MoodGroup/actigraphy/KMOA/raw'
olddir= '/data/MoodGroup/actigraphy/KMOA_old'
#for each subject:
for subj in dat.iterrows():
    #if there is an actigraphy ID
        if not pd.isnull(subj[1]['act_id']):
            #make them a folder
            #check for an AWD file
            #in KMOA_old
            check = glob(os.path.join(olddir,'data',subj[1]['act_id']+'-KMOA','*AWD'))
            #if there was an awd there, keep it
            if len(check) == 0:
                check = glob(os.path.join(main_dir,'wally_old','*'+subj[1]['act_id'])+'*.AWD')
            if len(check) == 0:
                check = glob(os.path.join(main_dir,'wally_raw_data',subj[1]['act_id']+'.AWD'))
            #ok, now we have a "check" which is 0 if there's nothing and not 0 if there's something. so
            new['num_awds'].append(len(check))
        else:
            new['num_awds'].append(0)
```

```python
awd_cols = []
for i in range(0,max(new['num_awds'])):
    mylist = []
    mylist.append('awd_gender_' + str(i))
    mylist.append('awd_age_' + str(i))
    mylist.append('awd_date_match_' + str(i))
    awd_cols.append(mylist)

for mylist in awd_cols:
    for col_name in mylist:
        new[col_name] = []
awd_cols
```

Ok, now let's go in and actually move and check everything

```python
rawdir= '/data/MoodGroup/actigraphy/KMOA/raw'
olddir= '/data/MoodGroup/actigraphy/KMOA_old'
#for each subject:
for subj in dat.iterrows():
    #if there is an actigraphy ID
        if not pd.isnull(subj[1]['act_id']):
            sub_id = 'subMOA-' + str(subj[1]['sub_no'])
            subdir = os.path.join(rawdir,sub_id)
            origdir = os.path.join(subdir,'orig')
            os.mkdir(subdir)
            os.mkdir(origdir)
```

```python
rawdir= '/data/MoodGroup/actigraphy/KMOA/raw'
olddir= '/data/MoodGroup/actigraphy/KMOA_old'
#for each subject:
for subj in dat.iterrows():
    #if there is an actigraphy ID
        if not pd.isnull(subj[1]['act_id']):
            sub_id = 'subMOA-' + str(subj[1]['sub_no'])
            subdir = os.path.join(rawdir,sub_id)
            origdir = os.path.join(subdir,'orig')
            #check for an AWD file
            #in KMOA_old
            check = glob(os.path.join(olddir,'data',subj[1]['act_id']+'-KMOA','*AWD'))
            #if there was an awd there, keep it
            if len(check) == 0:
                check = glob(os.path.join(main_dir,'wally_old','*'+subj[1]['act_id'])+'*.AWD')
            #if you still haven't found an awd check wally_raw_data
            if len(check) == 0:
                check = glob(os.path.join(main_dir,'wally_raw_data',subj[1]['act_id']+'.AWD'))
            #ok, now we have a "check" which is 0 if there's nothing and not 0 if there's something. so
            #now for each file we want to move it over, read it, and get the gender, age, and see if the date of infusion is in the file
            for idx,file in enumerate(check):
                n = str(idx)
                fn = (os.path.split(file)[-1])
                #move with original name to orig
                shutil.copyfile(file,os.path.join(origdir,fn))
                #rename and move to subdir
                shutil.copyfile(file,os.path.join(subdir,sub_id + '_' + n +'.AWD'))
                #read AWD
                #get gender, age
                try:
                    awd_dat = act.read_AWD(file)
                    new['awd_gender_' + n].append(awd_dat['hdr']['gender'])
                    new['awd_age_' + n].append(awd_dat['hdr']['age'])

                    # now check to see if date matches
                    if not pd.isnull(subj[1]['infusion']):
                        info = subj[1]['infusion'].split('/')
                        dt= datetime(year = int(info[2]), month = int(info[0]), day = int(info[1]), hour=9)

                        new['awd_date_match_' + n].append(dt in awd_dat['dt_list'])
                    else:
                        new['awd_date_match_' + n].append("")
                except:
                    new['awd_gender_' + n].append('')
                    new['awd_age_' + n].append('')
                    new['awd_date_match_' + n].append('')
                    print("could not read " + file)
                    
            #add empties to leftover entries
            for idx in range(len(check),3):
                n = str(idx)
                new['awd_gender_' + n].append('')
                new['awd_age_' + n].append('')
                new['awd_date_match_' + n].append('')
        else:
            for idx in range(0,3):
                n = str(idx)
                new['awd_gender_' + n].append('')
                new['awd_age_' + n].append('')
                new['awd_date_match_' + n].append('')
                
```

```python
for item in new.keys():
    dat[item] = new[item]
```

```python
dat.to_csv(os.path.join(main_dir,"KMOA_summary.csv"))
```

OK now to move over and document the sleep logs and calendars

```python
cals = []
rawdir= '/data/MoodGroup/actigraphy/KMOA/raw'
olddir= '/data/MoodGroup/actigraphy/KMOA_old'
for subj in dat.iterrows():
    sub_id = 'subMOA-' + str(subj[1]['sub_no'])
    subdir = os.path.join(rawdir,sub_id)
    num = str(subj[1]['sub_no'])
    check = glob(os.path.join(main_dir,'calendars_30Jul2019',num+'*'))
    if len(check)>0:
        for file in check:
            fn = (os.path.split(file)[-1])
            shutil.copyfile(file,os.path.join(subdir,fn))
        cals.append(1)
    else:
        cals.append(0)
            
```

Finally get whatever sleep logs there may be

```python
logs = []
for subj in dat.iterrows():
    if not pd.isnull(subj[1]['act_id']):
        sub_id = 'subMOA-' + str(subj[1]['sub_no'])
        subdir = os.path.join(rawdir,sub_id)
        check = glob(os.path.join(olddir,'sleeplogs',subj[1]['act_id'])+'*')
        if len(check)>0:
            for file in check:
                fn = (os.path.split(file)[-1])
                shutil.copyfile(file,os.path.join(subdir,fn))
            logs.append(1)
        else:
            logs.append(0)
    else:
        logs.append(0)
```

```python
rawdir= '/data/MoodGroup/actigraphy/KMOA/raw'
outdir = '/data/MoodGroup/actigraphy/KMOA/derivatives/old_proc/'
processed = []

for subj in dat.iterrows():
    if not pd.isnull(subj[1]['act_id']):
        sub_id = 'subMOA-' + str(subj[1]['sub_no'])
        sub_out = os.path.join(outdir,sub_id)
        check = glob(os.path.join(sub_out,'%s_processed*xlsx'%subj[1]['act_id']))
        if len(check)==1:
            fn_full = check[0]
            fn = os.path.split(fn_full)[-1]
            if len(fn.split("_")) == 2:
                skiprows = [0,1,2,3,4]
            elif len(fn.split("_")) == 3:
                skiprows = [0,1]
            tmp = pd.read_excel(fn_full,skiprows=skiprows,names=['day','time','activity'])
            tmp.to_csv(os.path.join(sub_out,'%d_processed.csv'%subj[1]['sub_no']))
```
