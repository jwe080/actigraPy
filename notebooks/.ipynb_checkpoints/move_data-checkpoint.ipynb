Move stuff from wally folder

```python
import pandas as pd
import os
import shutil
from glob import glob
import numpy as np
```

```python
main_dir = '/data/MoodGroup/actigraphy/KMOA'
```

```python
dat = pd.read_csv(os.path.join(main_dir,"KMOA_subjs.csv"))
```

```python
for row in dat.iterrows():
    sub_dir = os.path.join(main_dir,'raw','subMOA-' + str(int(row[1]['sub_no'])))
    orig_dir = os.path.join(sub_dir, 'orig')
    fn_awd = os.path.join(main_dir,'wally_raw_data',row[1]['act_id']+'.AWD')
    if os.path.isfile(fn_awd):
        shutil.copyfile(fn_awd,os.path.join(orig_dir,row[1]['act_id']+'.AWD'))
        shutil.copyfile(fn_awd,os.path.join(sub_dir,'subMOA-' + str(int(row[1]['sub_no']))+'.AWD'))

        
        
```

Move stuff from old KMOA folder

```python
old_dir = '/data/MoodGroup/actigraphy/KMOA_old/data'
```

```python
for row in dat.iterrows():
    sub_dir = os.path.join(main_dir,'raw','subMOA-' + str(int(row[1]['sub_no'])))
    orig_dir = os.path.join(sub_dir, 'orig')

    act_id = row[1]['act_id']
    #check if there is a subj file
    fns = (glob(os.path.join(old_dir,'*'+act_id+'*')))
    
    if len(fns) > 0:
        fn = fns[0]
        #check if, for that subj file, there is data
        fn_awds = glob(os.path.join(fn,'*.AWD'))
        if len(fn_awds) > 0:
            print(len(fn_awds))
            fn_awd = fn_awds[0]
            print(fn_awd)
            shutil.copyfile(fn_awd,os.path.join(orig_dir,row[1]['act_id']+'.AWD'))
            shutil.copyfile(fn_awd,os.path.join(sub_dir,'subMOA-' + str(int(row[1]['sub_no']))+'.AWD'))
```
